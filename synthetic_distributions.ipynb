```python
from sklearn import datasets
from matplotlib import pyplot as plt

circles, labels = datasets.make_circles(1000, noise=0.05, factor=0.6, shuffle=True)
circles = circles.T
plt.scatter(circles[0]*5, circles[1]*10)
```

```python
make_moons, labels = datasets.make_moons(n_samples=1000, noise=0.05)

make_moons = make_moons.T
plt.scatter(make_moons[0]*5, make_moons[1]*10)
```

```python
import numpy

complex_data = numpy.hstack([make_moons + 2, circles ]) 

plt.scatter(complex_data[0]*5, complex_data[1]*10)
```
